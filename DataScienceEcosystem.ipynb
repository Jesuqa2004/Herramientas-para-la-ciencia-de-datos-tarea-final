```python
def minutes_to_hours(minutes):
  """Converts minutes to hours.

  Args:
    minutes: The number of minutes to convert.

  Returns:
    The equivalent number of hours.
  """
  hours = minutes / 60
  return hours

amount_minutes = 581
amount_hours = minutes_to_hours(amount_minutes)
print(f"{amount_minutes} minutes is equivalent to {amount_hours} hours.")
```
